```python
import kagglehub
import pandas as pd
import matplotlib.pyplot as plt
import os

# Download latest version
path = kagglehub.dataset_download("uciml/pima-indians-diabetes-database")

print("Path to dataset files:", path)
```

### Load Data and Initial Inspection

Now, I will load the `diabetes.csv` file into a pandas DataFrame and display its first 5 rows, information, and initial missing value counts.

```python
# Construct the full file path
file_name = 'diabetes.csv'
file_path = os.path.join(path, file_name)

df = pd.read_csv(file_path)

print("First 5 rows of the DataFrame:")
display(df.head())
print("\nDataFrame Info:")
df.info()
print("\nInitial missing values:")
print(df.isnull().sum())
```

### Handle Zero Values

I will now replace '0' values with `pd.NA` in the 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', and 'BMI' columns, and then check the missing value counts again.

```python
cols = ["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]
df[cols] = df[cols].replace(0, pd.NA)

print("Missing values after replacing 0s with pd.NA:")
print(df.isnull().sum())
```

### Visualize Data Distribution

Finally, I will visualize the distribution of the 'Glucose' and 'Age' columns using a histogram and a box plot, respectively.

```python
plt.figure(figsize=(10, 6))
plt.hist(df["Glucose"].dropna(), bins=20, edgecolor='black')
plt.title('Distribution of Glucose')
plt.xlabel('Glucose Level')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df["Age"].dropna())
plt.title('Distribution of Age')
plt.ylabel('Age')
plt.grid(axis='y', alpha=0.75)
plt.show()
```
