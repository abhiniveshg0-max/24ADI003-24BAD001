```python
import kagglehub
import os
import pandas as pd

# Download the latest version of the dataset
path = kagglehub.dataset_download("debashis74017/lic-stock-price-data")

print("Dataset path:", path)

# List files to see available data
files = os.listdir(path)
print("Files in dataset:", files)

# Identify the CSV file
csv_file = next((f for f in files if f.endswith('.csv')), None)

if csv_file:
    full_path = os.path.join(path, csv_file)
    df = pd.read_csv(full_path)
    print(f"\nSuccessfully loaded: {csv_file}")
    display(df.head())
else:
    print("\nNo CSV file found in the downloaded dataset.")
```
