```python
import kagglehub

# Download latest version of the dataset
path = kagglehub.dataset_download("harishkumardatalab/housing-price-prediction")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# The 'path' variable from the previous cell contains the directory where the dataset was downloaded.
# We assume the CSV file is named 'Housing.csv' inside that directory.
csv_file_path = os.path.join(path, "Housing.csv")
df = pd.read_csv(csv_file_path)

print(df.head())
```

```python
print(df.isnull().sum())
```

```python
plt.scatter(df["area"], df["price"])
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Area vs. Price")
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.select_dtypes("number").corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix of Numerical Features")
plt.show()
```
